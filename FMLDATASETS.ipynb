```python
cols = list(df.columns[4:])
cols
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scaled_data = scalar.fit_transform(df[cols[:-1]].to_numpy())
scaled_df = pd.DataFrame(scaled_data, columns=cols[:-1])
scaled_df
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scaled_df[cols[-1]] = scalar.fit_transform(df[cols[-1]].to_numpy().reshape(-1, 1)).reshape(-1)
scaled_df
```

```python
def get_class(ghi):
    if ghi > 0.4:
        return 'running'
    elif ghi >= 0.4:
        return 'monitoring'
    elif ghi >= -1:
        return 'inspecting'
        
```

```python
scaled_df['class'] = scaled_df['GHI (Wh/m2)'].apply(get_class)
scaled_df
```
